```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("A5.csv")
df
```

```python
# Ans 1 (a).
# Removing all the records where Rating is not present
df4 = df.dropna(subset=['Rating'])
df4
```

```python
# Ans 1(b):
# Checking for null values in android version.
df4[pd.isnull(df['Android Ver'])]
```

```python
# Filling the remaining missing values with the mode and dropping the third value.
df4=df4.drop([10472])
```

```python
# Replacing the Android version with mode
df4['Android Ver'].fillna('4.1 and up',inplace=True)
```

```python
# Relacing the current version with the most common value
df4[pd.isnull(df4['Current Ver'])]
```

```python
df4['Current Ver'].mode()
```

```python
df4['Current Ver'].fillna('Varies with device',inplace=True)
```

```python
# Remove "$" sign from price column
df4['Price']=df4['Price'].str.replace('$','')
df4['Price']=pd.to_numeric(df4['Price'])
```

```python
# Removing '+' and ',' from Installs column 
df4['Installs']=df4['Installs'].str.replace(',','').str.replace('+','')
df4['Installs']=pd.to_numeric(df4['Installs'])
```

```python
# Checking if all the ratings are between 1 and 5
print(df4[(df4['Rating']<1)])
print(df4[(df4['Rating']>5)])
```

```python
# Identifying the outliers through Boxplot
df4.boxplot(column=['Price'],grid=False)
```

```python
# Dropping all the values grater than 299
df4.drop(df4[df4['Price']>30].index,inplace=True)
```

```python
df4.boxplot(column=['Price'],grid=False)
```

```python
df4['Reviews']=pd.to_numeric(df4['Reviews'])
```

```python
# Boxplot for Reviews
df4.boxplot(column=['Reviews'],grid=False)
```

```python
# Limiting the Rows with Reviews<1000000 
df_reviews=df4[df4['Reviews']<1000000]
df_reviews
```

```python
df_reviews.boxplot(column=['Reviews'],grid=False)
```

```python
# Dropping the values with more than 1000000 reviews 
df4.drop(df4[df4['Reviews']>1000000].index,inplace=True)
```
